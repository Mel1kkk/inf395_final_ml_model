```python
import requests
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
print("Data Scraping")
```

```python
for p in range(1, 90):
    print(p)
    
    url = f"https://www.metacritic.com/browse/game/?releaseYearMin=1958&releaseYearMax=2024&page={p}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    r = requests.get(url, headers=headers)

    sleep(10)
    
    soup = BeautifulSoup(r.text, "lxml")
    
    games = soup.findAll("div", class_="c-finderProductCard")
    
    for game in games:
         
        game_name = game.find("div", class_="c-finderProductCard_title").find_all("span")[1].text.strip()
         
        place_in_top = game.find("h3", class_="c-finderProductCard_titleHeading").find("span").text.strip()
         
        game_link = "https://rawg.io" + game.find("a", href=True)["href"]
        game_link = game_link.rstrip('/')
        game_link = game_link.replace('/game/','/games/')
        game_response = requests.get(game_link, headers=headers, timeout=10)  # Add timeout here too
        rawg = BeautifulSoup(game_response.text, 'lxml')
        m = rawg.find(class_ = "game-content-columns")

        
        release_date = game.find("div", class_="c-finderProductCard_meta").find_all("span")[0].text.strip()

        meta_spans = game.find("div", class_="c-finderProductCard_meta").find_all("span")
        rating_letter = meta_spans[2].text.strip() if len(meta_spans) > 2 else "Not rated"
        
         # rating_letter = game.find("div", class_="c-finderProductCard_meta").find_all("span")[2].text.strip()
        
        metascore_rating = game.find("div", class_="c-finderProductCard_metascoreValue").find("span").text.strip()




        platforms = np.nan
        genre = np.nan
        developers = np.nan
        publisher = np.nan
        age_restriction = np.nan
        other_games_in_series = np.nan
        tags = np.nan
        website = np.nan
        rating_in_rawg = np.nan
        rating_in_a_year_rawg = np.nan
        
        try:
            avg_playtime = m.find(class_ = "game__head").find(class_ = "game__meta-playtime").text.strip()
        except AttributeError:
            avg_playtime = np.nan

        
        try: 
            main = m.find(class_ = "game__meta").find_all(class_ = "game__meta-block")



            try:
                rating_in_rawg = m.find_all(class_ = "rating-chart__chart")[1].text.strip()
            except IndexError:
                rating_in_rawg = np.nan

            try:
                rating_in_a_year_rawg = m.find_all(class_ = "rating-chart__chart")[2].text.strip()
            except IndexError:
                rating_in_a_year_rawg = np.nan
        
            for block in main:
            
                title = block.find(class_="game__meta-title").text.strip()

    
                if title == "Platforms":
                    platforms = block.find(class_="game__meta-text").text.strip()
                elif title == "Genre":
                    genre = block.find(class_="game__meta-text").text.strip()
                elif title == "Developer":
                    developers = block.find(class_="game__meta-text").text.strip()
                elif title == "Publisher":
                    publisher = block.find(class_="game__meta-text").text.strip()
                elif title == "Age rating":
                    age_restriction = block.find(class_="game__meta-text").text.strip()
                elif title == "Other games in the series":
                    other_games_in_series = block.find(class_="game__meta-text").text.strip()
                elif title == "Tags":
                    tags = block.find(class_="game__meta-text").text.strip()
                elif title == "Website":
                    website = block.find(class_="game__meta-website").text.strip()


        except AttributeError:
            main = np.nan
        game_data.append([game_name, place_in_top, game_link, release_date, rating_letter, metascore_rating, avg_playtime, rating_in_rawg,rating_in_a_year_rawg, platforms, genre, developers, publisher, age_restriction, other_games_in_series,tags,website])
```

```python
header = (["game_name", "place_in_top", "game_link", "release_date", "rating_letter", "metascore_rating",'avg_playtime', 'rating_in_rawq', 'rating_in_that_year', 'platforms', 'genre', 'developers', 'publisher', 'age_restriction', 'other_games_in_series','tags','website'])
```

```python
df = pd.DataFrame(game_data, columns=header)
df.to_csv('game_data.csv', index = False, sep=',')
```

```python
print("Data Cleaning")
```

```python
df2 = df2.drop_duplicates()
```

```python
threshold = 11
df2 = df2[df2.isnull().sum(axis=1) < threshold]
```

```python
df2['avg_playtime'] = df2['avg_playtime'].str.strip('Average Playtime: ')
```

```python
df2['avg_playtime'] = df2['avg_playtime'].str.strip('hours')
```

```python
df2['avg_playtime'] = pd.to_numeric(df2['avg_playtime'], errors='coerce')
```

```python
df2['avg_playtime'] = df2['avg_playtime'].fillna(df2['avg_playtime'].mean())
df2['avg_playtime'].head()
```

```python
df2[['rating_in_rawq', 'genre_of_rating']] = df2['rating_in_rawq'].str.extract(r'#(\d+)([A-Za-z]+)')
```

```python
df2['rating_in_rawq'] = pd.to_numeric(df2['rating_in_rawq'], errors='coerce')
```

```python
df2['rating_in_rawq'] = df2['rating_in_rawq'].fillna(df2['rating_in_rawq'].max() + 1)
```

```python
df2['genre_of_rating'] = df2['genre_of_rating'].fillna('Not in a rating')
```

```python
df2 = df2[df2['platforms'].notnull()]
```

```python
df2['rating_in_that_year'] = df2['rating_in_that_year'].str.extract(r'#(\d+)')
df2['rating_in_that_year']
```

```python
df2['rating_in_that_year'] = pd.to_numeric(df2['rating_in_that_year'], errors='coerce')
```

```python
df2['rating_in_that_year'] = df2['rating_in_that_year'].fillna(df2['rating_in_that_year'].max()+ 1)
df2['rating_in_that_year']
```

```python
df2[['developers','publisher','tags','website', 'genre']] = df2[['developers','publisher','tags','website', 'genre']].fillna('Not defined')
```

```python
df2['other_games_in_series'] = df2['other_games_in_series'].fillna('No')
```

```python
replace_dict = {
    '13+ Teen': '13',
    'Not rated': '0',
    '17+ Mature': '17',
    '10+': '10',
    '0+ Everyone': '0',
    '1+ Everyone': '1',
    '10+ Everyone 10+': '10',
    '18+ Adults Only': '18',
    'RP Rating Pending': '0'
}
```

```python
df2['age_restriction'] = df2['age_restriction'].replace(replace_dict)
df2['age_restriction'] = df2['age_restriction'].astype(int)
```

```python
path_final = "cleaned_game_data.csv"
df_final = pd.read_csv(path_final)

desired_order = [
    "game_name", "place_in_top", "genre", "release_date", 
    "rating_letter", "age_restriction", "avg_playtime", "platforms", "developers", 
    "publisher", "other_games_in_series", "tags", "metascore_rating", 
    "rating_in_that_year", "rating_in_rawq", "genre_of_rating", "game_link", "website"
]

df_final = df_final[desired_order]

rename_columns = {
    "genre": "genres",
    "rating_letter": "rated_age_letter",
    "rating_in_that_year": "place_in_that_year",
    "rating_in_rawq": "place_in_rawq_by_genre",
    "genre_of_rating": "genre_of_this_place"
}

df_final.rename(columns=rename_columns, inplace=True)

df_final['place_in_top'] = df_final['place_in_top'].str.replace(',', '').astype(float).round().astype(int)
df_final['place_in_that_year'] = df_final['place_in_that_year'].astype(float).round().astype(int)
df_final['place_in_rawq_by_genre'] = df_final['place_in_rawq_by_genre'].astype(float).round().astype(int)
df_final['avg_playtime'] = df_final['avg_playtime'].round(1)


df_final.to_csv(path_final, index=False, sep=',')
df_final.tail(5)
```
